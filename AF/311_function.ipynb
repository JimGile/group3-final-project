```python
def classify_issue(description):
    topics_dict
    for keyword, category in topics_dict.items():
        if keyword in description.lower():
            return {
                "category": category,
                "url": category_links[category]
            }
    return {
        "category": "Other",
        "url": category_links["Other"]
    }

# Example usage
issues_description = "There is graffiti on the wall near the subway station."
result = classify_issue(issues_description)
print(result)
```
