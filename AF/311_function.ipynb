```python
def classify_issue(topics, categories):
    keywords = [topic.strip() for topic in topics.split(',')]
    issues_responses = []
    for keyword, details_dict in categories.items():
        if keyword in topics:
            issues_responses.append(
            {
                "category": keyword,
                "link": details_dict['link'],
                "response": details_dict['response']
            }
            )
    return issues_responses

# Example usage
topics = "Police, Policy"
result = classify_issue(topics, categories)
print(result)
```
